```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("upi_data_enhanced.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Month'] = pd.to_datetime(df['Month'], format='%b-%y')
df = df.sort_values('Month')
```

```python
df = df.rename(columns={
    'No. of Banks live on UPI': 'Banks_Live',
    'Volume (in Mn)': 'Txn_Volume_Mn',
    'Value (in Cr.)': 'Txn_Value_Cr'
})
```

```python
df.columns
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df = df.where(pd.notnull(df), None)
```

```python
df.to_csv("cleaned_upi_data.csv", index=False)
```

```python
engine = create_engine(
    "mysql+pymysql://root:pricass00@localhost:3306/fintech_upi"
)
```

```python
engine.connect()
print("MySQL connection successful")
```

```python
df.to_sql(
    name="upi_transactions",
    con=engine,
    if_exists="replace",
    index=False
)

print("Data successfully loaded into MySQL")
```
